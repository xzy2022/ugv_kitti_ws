## 1. 读取二进制数据与数据分析

### a. 数据分布

读取、分析与可视化都是00000.bin帧的数据

```python
def task_qualified_read(file_path):
    """
    读取 KITTI .bin 文件，返回原始点云矩阵
    """
    # 1. 使用 numpy 的 fromfile 读取二进制流
    # KITTI 数据全部是 float32 格式 [cite: 59]
    scan = np.fromfile(file_path, dtype=np.float32)
    
    # 2. 重塑矩阵：每4个数据代表一个点 (x, y, z, intensity) [cite: 61]
    # reshape(-1, 4) 会自动计算点的数量 N
    points = scan.reshape((-1, 4))
    
    return points

# --- 测试运行 ---
test_file = os.path.join(VELODYNE_PATH, '000000.bin')
points = task_qualified_read(test_file)

print(f"读取完成。")
print(f"点云形状: {points.shape}") # 应该看到类似 (12xxxx, 4)
print(f"前5个点数据:\n{points[:5, :]}")
```

```python
import pandas as pd
import numpy as np

def analyze_stats(points):
    """
    计算点云的统计特性
    输入: points (N, 4) 矩阵
    """
    # 将数据转换为 DataFrame 方便查看，命名列
    df = pd.DataFrame(points, columns=['x', 'y', 'z', 'reflectance'])
    
    # 计算距离 (Range)，这是雷达很重要的属性
    df['range_2d'] = np.sqrt(df['x']**2 + df['y']**2)
    df['range_3d'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)
    
    print("====== 点云基础统计信息 ======")
    print(df.describe().round(3)) # 保留3位小数
    
    print("\n====== Reflectance 零值分析 ======")
    # 1. 严格等于 0 的点
    exact_zero_count = (df['reflectance'] == 0).sum()
    print(f"Reflectance == 0 的点数: {exact_zero_count}")
    
    # 2. 近似等于 0 的点 (处理浮点数精度，或者非常小的噪声值)
    # 这里的阈值设为 1e-6，你可以根据需要调整，例如之前可视化的 0.001
    epsilon = 1e-6
    approx_zero_count = (df['reflectance'] <= epsilon).sum()
    print(f"Reflectance ≈ 0 (<= {epsilon}) 的点数: {approx_zero_count}")
    
    return df

# --- 使用示例 ---
df = analyze_stats(points)
```

### b. 雷达采样顺序

验证可得，雷达是顺时针旋转，从正前方x轴正方向开始

```python
import numpy as np
import matplotlib.pyplot as plt
import os

def verify_scan_order(file_path):
    """
    验证点云是否按照旋转顺序排列
    """
    # 1. 读取数据
    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    
    # 2. 提取 X, Y 并计算水平方位角 (Azimuth)
    # arctan2 的范围是 [-pi, pi]
    x = points[:, 0]
    y = points[:, 1]
    azimuth = np.arctan2(y, x)
    
    # 3. 可视化分析
    # 为了看清细节，我们只取前 2000 个点和中间的一部分点
    plt.figure(figsize=(12, 6))
    
    # 子图1：宏观趋势 (前 10% 的点)
    limit = int(len(azimuth) * 0.1)
    plt.subplot(1, 2, 1)
    plt.plot(azimuth[:limit], '.', markersize=0.5)
    plt.title(f"Azimuth vs Index (First {limit} points)")
    plt.xlabel("Point Index")
    plt.ylabel("Azimuth (rad)")
    plt.grid(True)
    
    # 子图2：微观细节 (前 200 个点)
    # Velodyne 64线雷达是垂直方向同时（或极短时间内）发射多束激光
    # 因此我们可能会看到一簇点的角度非常接近，然后才变化
    plt.subplot(1, 2, 2)
    plt.plot(azimuth[:200], '.-')
    plt.title("Micro View (First 200 points)")
    plt.xlabel("Point Index")
    plt.ylabel("Azimuth (rad)")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # 4. 数值验证：计算角度的差分
    # 注意：由于是在圆周上旋转，角度会在 pi 和 -pi 之间跳变，需要处理
    # 简单的验证方法是看局部是否大概率单调递增
    diffs = np.diff(azimuth)
    # 过滤掉 wrap-around (例如从 3.14 跳变到 -3.14 的大跳变)
    valid_diffs = diffs[np.abs(diffs) < 5.0] 
    
    positive_ratio = np.sum(valid_diffs > 0) / len(valid_diffs)
    print(f"=== 验证结果 ===")
    print(f"正向角度变化比例: {positive_ratio:.2%}")
    if positive_ratio > 0.9:
        print("结论: 点云是按逆时针旋转顺序存储的 (符合预期)")
    elif positive_ratio < 0.1:
        print("结论: 点云是按顺时针旋转顺序存储的")
    else:
        print("结论: 点云顺序可能是杂乱的或经过了重排")

# --- 请替换为您本地的真实路径 ---
test_file = os.path.join(VELODYNE_PATH, '000000.bin') 
verify_scan_order(test_file)
```

## 2. 数据分布可视化

### a.地面高度分析

观察z=-1.73(理想地面)为界限的数据分布是否有明显不同。结果证明，没有。

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

# 正常高度的颜色渐进 colormap（你也可以换）
KITTICMAP = LinearSegmentedColormap.from_list(
    "kitti_height", ["#0000ff", "#00ff00", "#ff0000"]
)

def visualize_spatial(df, z_split=-1.73):
    """
    左图（BEV）：
        - 空白背景 = 黑色
        - z < z_split = 浅蓝色
        - z >= z_split = 灰度随 z 增强变亮
    右图：
        - 比较 z < z_split 与 z >= z_split 的 reflectance 直方图分布
    """

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    # ============================================================
    # 1. 左图 BEV
    # ============================================================
    x = df['x'].values
    y = df['y'].values
    z = df['z'].values

    z_min = z.min()
    z_max = z.max()

    colors = np.zeros((len(df), 4))  # RGBA

    # --- ① 阈值以下：浅蓝色 ---
    low_mask = (z < z_split)
    colors[low_mask] = [0.4, 0.6, 1.0, 1.0]

    # --- ② 阈值以上：亮度随高度增强 ---
    high_mask = (z >= z_split)
    z_norm = (z[high_mask] - z_split) / (z_max - z_split)
    z_norm = np.clip(z_norm, 0, 1)

    min_gray = 0.3
    gray_vals = min_gray + z_norm * (1.0 - min_gray)  # 0.3 ~ 1.0 之间

    colors[high_mask] = np.stack([gray_vals, gray_vals, gray_vals, np.ones_like(gray_vals)], axis=1)

    # 设置背景为黑色
    ax1.set_facecolor("black")

    ax1.scatter(x, y, c=colors, s=0.1)
    ax1.axis("equal")
    ax1.set_title("Bird's Eye View (Custom Color Scheme)")
    ax1.set_xlabel("X (Forward)")
    ax1.set_ylabel("Y (Left/Right)")

    # ============================================================
    # 2. 右图 Reflectance 直方图对比
    # ============================================================

    refl_low = df.loc[df['z'] < z_split, 'reflectance']
    refl_high = df.loc[df['z'] >= z_split, 'reflectance']

    bins = 50

    ax2.hist(
        refl_low,
        bins=bins,
        alpha=0.6,
        color='skyblue',
        label=f"z < {z_split:.2f}",
        density=False
    )
    ax2.hist(
        refl_high,
        bins=bins,
        alpha=0.6,
        color='white',
        edgecolor='gray',
        label=f"z ≥ {z_split:.2f}",
        density=False
    )

    ax2.set_title("Reflectance Distribution Comparison")
    ax2.set_xlabel("Reflectance")
    ax2.set_ylabel("Density")
    ax2.legend()

    plt.show()


# --- 运行可视化 ---
visualize_spatial(df, z_split=-1.73)
```

### b. 地面坡度分析

可以明显看出这是一个上坡的地形，110m的长度上坡1.3的高度，坡度角大约0.677°

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_ground_slope(df, z_split=-1.73):
    """
    基于 X-Z 侧视图的地面坡度可视化分析。
    
    参数
    ----
    df : DataFrame
        必须包含 'x', 'y', 'z', 'reflectance'
    z_split : float
        视为地面高度，用于可视化参考基准（可选）
    """

    plt.figure(figsize=(18, 6))

    # 仅使用 Y 轴中间区域，减少两侧障碍物干扰（你之前的设计保持）
    mask = (df['y'] > -5) & (df['y'] < 5)
    subset = df[mask]

    sc = plt.scatter(
        subset['x'], subset['z'],
        s=0.1,
        c=subset['reflectance'],
        cmap='jet'
    )

    plt.title("Ground Slope Visualization (Side View: X vs Z)")
    plt.xlabel("X (Forward)")
    plt.ylabel("Z (Height)")

    # 限制Z范围以聚焦于地面高度
    plt.ylim(-3, 1)

    # 画出地面基准线
    plt.axhline(z_split, color='white', linestyle='--', linewidth=1, alpha=0.7, 
                label=f"Ground Reference (z={z_split})")

    plt.legend()
    plt.colorbar(sc, label='Reflectance')

    plt.show()

visualize_ground_slope(df)
```

### c. 强度的分布分析

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_distributions_manual(df, vline1_pos=0.25, vline2_pos=0.35, n_bins_main=60, n_bins_zoom=50):
    """
    可视化强度分布直方图。
    图1：整体分布图，并绘制两条可调整位置的竖线。
    图2：聚焦于 0 到 0.03 范围内的分布图。
    
    参数:
    - df: 包含 'reflectance' 列的 DataFrame。
    - vline1_pos: 第一条竖线的位置 (默认 0.25)。
    - vline2_pos: 第二条竖线的位置 (默认 0.35)。
    - n_bins_main: 主图的直方图 bin 数量。
    - n_bins_zoom: 放大图的直方图 bin 数量。
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    refl = df['reflectance'].values
    
    # =========================
    # 1. 主直方图 + 手动竖线
    # =========================
    # 绘制整体分布直方图
    ax1.hist(refl, bins=n_bins_main, log=False, color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_title("Reflectance Distribution (Main View)")
    ax1.set_xlabel("Reflectance")
    ax1.set_ylabel("Count")
    
    # 绘制指定的两条竖线
    # 使用不同的颜色和线型来区分
    ax1.axvline(x=vline1_pos, color='red', linestyle='--', linewidth=2, label=f'Manual Split 1: {vline1_pos}')
    ax1.axvline(x=vline2_pos, color='green', linestyle='-.', linewidth=2, label=f'Manual Split 2: {vline2_pos}')
    
    # 添加图例以显示竖线的值
    ax1.legend()
    
    # =========================
    # 2. 聚焦直方图 (0 - 0.03 范围)
    # =========================
    # 定义聚焦范围
    zoom_min, zoom_max = 0.0, 0.03
    
    # 筛选出该范围内的数据
    refl_zoom = refl[(refl >= zoom_min) & (refl <= zoom_max)]
    
    if len(refl_zoom) > 0:
        # 绘制放大后的直方图
        # 使用较细的 bin 来展示该狭窄范围内的细节
        ax2.hist(refl_zoom, bins=n_bins_zoom, color='orange', edgecolor='black', alpha=0.7)
    else:
        print(f"Warning: No data found in range [{zoom_min}, {zoom_max}]")
        
    ax2.set_title(f"Reflectance Distribution (Zoomed: {zoom_min}-{zoom_max})")
    ax2.set_xlabel("Reflectance")
    ax2.set_ylabel("Count")
    
    # 强制设置X轴范围，确保视图聚焦
    ax2.set_xlim(zoom_min, zoom_max)
    # 添加网格方便查看
    ax2.grid(True, axis='x', linestyle='--', alpha=0.5)
    
    plt.tight_layout()
    plt.show()


# 使用默认参数运行 (竖线在 0.25 和 0.35)
print("Displaying with custom split lines (0.25, 0.35)...")
visualize_distributions_manual(df, vline1_pos=0.25, vline2_pos=0.35)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_reflectance_subsets(df, range_min=0.25, range_max=0.35, z_split=-1.73, point_size=0.5):
    """
    可视化两类特定反射率的点云分布（鸟瞰图 BEV）。
    两张图均遵循统一的视觉标准：
        - 背景 = 黑色
        - z < z_split = 浅蓝色 (地面)
        - z >= z_split = 灰度随 z 增强变亮 (障碍物)
    
    参数:
    - df: 包含 'x', 'y', 'z', 'reflectance' 的 DataFrame
    - range_min, range_max: 右图筛选反射率的范围
    - z_split: 地面与障碍物的高度分割阈值 (默认 -1.73)
    - point_size: 散点大小
    """
    
    # 获取全局最大高度，保证两张图的灰度亮度标准一致
    global_z_max = df['z'].max()
    
    # 定义一个内部辅助函数来生成颜色
    def get_colors(z_values):
        colors = np.zeros((len(z_values), 4)) # RGBA
        
        # --- ① 阈值以下：浅蓝色 ---
        low_mask = (z_values < z_split)
        colors[low_mask] = [0.4, 0.6, 1.0, 1.0]
        
        # --- ② 阈值以上：亮度随高度增强 ---
        high_mask = (z_values >= z_split)
        if np.any(high_mask):
            # 使用全局最大高度进行归一化，确保亮度在不同图中具有可比性
            denom = max(0.1, global_z_max - z_split) # 防止除以0
            z_norm = (z_values[high_mask] - z_split) / denom
            z_norm = np.clip(z_norm, 0, 1)
            
            min_gray = 0.3
            gray_vals = min_gray + z_norm * (1.0 - min_gray)
            
            # 赋值 RGBA (灰度 r=g=b)
            colors[high_mask] = np.stack([gray_vals, gray_vals, gray_vals, np.ones_like(gray_vals)], axis=1)
            
        return colors

    # 创建画布
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    # 提取原始数据
    x = df['x'].values
    y = df['y'].values
    z = df['z'].values
    refl = df['reflectance'].values

    # ============================================================
    # 1. 左图：只显示 Reflectance ≈ 0
    # ============================================================
    mask_zero = (refl <= 0.001)
    
    # 设置黑色背景
    ax1.set_facecolor("black")
    
    if np.any(mask_zero):
        x_sub = x[mask_zero]
        y_sub = y[mask_zero]
        z_sub = z[mask_zero]
        
        # 获取符合要求的颜色
        c_sub = get_colors(z_sub)
        
        ax1.scatter(x_sub, y_sub, c=c_sub, s=point_size)
        count_zero = np.sum(mask_zero)
    else:
        count_zero = 0
        
    ax1.axis("equal")
    ax1.set_title(f"Subset 1: Reflectance ≈ 0 (Count: {count_zero})\n(z < {z_split} Blue, z >= {z_split} Gray)")
    ax1.set_xlabel("X")
    ax1.set_ylabel("Y")
    ax1.grid(False)

    # ============================================================
    # 2. 右图：显示 Reflectance 在 [range_min, range_max] 之间
    # ============================================================
    mask_range = (refl >= range_min) & (refl <= range_max)
    
    # 设置黑色背景
    ax2.set_facecolor("black")
    
    if np.any(mask_range):
        x_sub = x[mask_range]
        y_sub = y[mask_range]
        z_sub = z[mask_range]
        
        # 获取符合要求的颜色
        c_sub = get_colors(z_sub)
        
        ax2.scatter(x_sub, y_sub, c=c_sub, s=point_size)
        count_range = np.sum(mask_range)
    else:
        count_range = 0
        
    ax2.axis("equal")
    ax2.set_title(f"Subset 2: Reflectance in [{range_min}, {range_max}] (Count: {count_range})\n(z < {z_split} Blue, z >= {z_split} Gray)")
    ax2.set_xlabel("X")
    ax2.set_ylabel("Y")
    ax2.grid(False)

    plt.tight_layout()
    plt.show()

# ==========================================
# 运行示例
# ==========================================
# 假设 df 已经存在
# z_split 默认为 -1.73，你可以根据需要调整
visualize_reflectance_subsets(df, range_min=0.25, range_max=0.35, z_split=-1.73)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

def visualize_distributions_trimodal(df, n_bins=60, sample_n=10000):
    """
    可视化强度分布，针对三峰分布（Tri-modal）进行分析。
    寻找三个主要波峰，并计算两个分割阈值。
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # =========================
    # 1. 强度直方图 + 多峰检测
    # =========================
    refl = df['reflectance'].values
    
    # 稍微增加 bin 的数量以获得更细致的轮廓，特别是针对 0 附近的窄峰
    counts, bin_edges = np.histogram(refl, bins=n_bins)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    
    # 画直方图
    ax1.hist(refl, bins=n_bins, log=False, color='skyblue', edgecolor='black', alpha=0.6)
    ax1.set_title("Reflectance Distribution (Tri-modal Analysis)")
    ax1.set_xlabel("Reflectance")
    ax1.set_ylabel("Count")
    
    # ---- 寻峰参数配置 ----
    # 针对 0 附近的尖峰，distance 不能太大，否则会漏掉紧挨着的峰
    # prominence 设为最大值的 2%，以捕捉可能较矮的中间峰
    min_distance = max(1, int(n_bins * 0.03)) 
    min_prominence = np.max(counts) * 0.02
    
    peaks_idx, properties = find_peaks(counts, distance=min_distance, prominence=min_prominence)
    
    splits = []
    sorted_peak_indices = []
    
    # 我们期望找到 3 个峰，但为了代码健壮性，我们处理 >= 2 的情况
    if len(peaks_idx) >= 2:
        # 1. 获取所有峰的高度
        peak_heights = counts[peaks_idx]
        
        # 2. 选取“最显著”的 Top 3 峰（按高度排序，取前3个）
        # 很多时候 0 处的峰很高，中间路面峰较矮，如果不取 Top N 可能会被噪声干扰
        top_peaks_zip = sorted(zip(peaks_idx, peak_heights), key=lambda x: x[1], reverse=True)[:3]
        
        # 3. 将这 3 个峰按“位置”（Reflectance大小）重新排序：左 -> 中 -> 右
        top_peaks_sorted = sorted(top_peaks_zip, key=lambda x: x[0])
        sorted_peak_indices = [p[0] for p in top_peaks_sorted]
        
        print("====== 三峰分析结果 ======")
        print(f"检测到的显著峰总数: {len(peaks_idx)}")
        print(f"选定的主要峰位置: {[round(bin_centers[i], 3) for i in sorted_peak_indices]}")
        
        # 4. 在相邻峰之间寻找谷底（分割点）
        colors = ['red', 'green', 'blue']
        labels = ['Peak 1 (Low)', 'Peak 2 (Mid)', 'Peak 3 (High)']
        
        # 绘制选定的峰
        for i, p_idx in enumerate(sorted_peak_indices):
            ax1.plot(bin_centers[p_idx], counts[p_idx], "o", 
                     color=colors[i], markersize=8, label=labels[i])
        
        # 计算分割线
        for i in range(len(sorted_peak_indices) - 1):
            p_left = sorted_peak_indices[i]
            p_right = sorted_peak_indices[i+1]
            
            # 在两个峰之间找最小值索引
            valley_region = counts[p_left:p_right+1]
            valley_rel_idx = np.argmin(valley_region)
            valley_idx = p_left + valley_rel_idx
            
            split_val = bin_centers[valley_idx]
            splits.append(split_val)
            
            ax1.axvline(split_val, color='black', linestyle='-.', linewidth=2, 
                        label=f'Split {i+1}: {split_val:.3f}')
            print(f"分割阈值 {i+1}: {split_val:.3f} (Between {bin_centers[p_left]:.2f} and {bin_centers[p_right]:.2f})")

        ax1.legend()
    else:
        print("未检测到足够的峰来进行多峰分割。")
        # 如果只有一个峰，就不画分割线了
        if len(peaks_idx) > 0:
             ax1.plot(bin_centers[peaks_idx], counts[peaks_idx], "ro")

    # =========================
    # 2. 距离 vs 强度 散点图（三类可视化）
    # =========================
    if len(df) > sample_n:
        sample = df.sample(n=sample_n, random_state=42)
    else:
        sample = df
        
    ax2.set_title("Reflectance vs Range-3D (Multi-class Split)")
    ax2.set_xlabel("Range-3D (m)")
    ax2.set_ylabel("Reflectance")
    ax2.grid(True, linestyle='--', alpha=0.3)
    
    # 根据分割点数量进行绘图
    if len(splits) == 0:
        ax2.scatter(sample['range_3d'], sample['reflectance'], s=3, alpha=0.5, c='gray')
    
    elif len(splits) == 1:
        # 只有两个类
        th1 = splits[0]
        mask1 = sample['reflectance'] < th1
        mask2 = sample['reflectance'] >= th1
        ax2.scatter(sample.loc[mask1, 'range_3d'], sample.loc[mask1, 'reflectance'], 
                    s=5, alpha=0.5, c='red', label='Class 1')
        ax2.scatter(sample.loc[mask2, 'range_3d'], sample.loc[mask2, 'reflectance'], 
                    s=5, alpha=0.5, c='blue', label='Class 2')
        ax2.axhline(th1, color='black', linestyle='-.')
        
    elif len(splits) >= 2:
        # 三个类 (或者更多，这里只处理前两个分割点形成的三类)
        th1 = splits[0]
        th2 = splits[1]
        
        # Class 1: < Threshold 1
        mask1 = sample['reflectance'] < th1
        # Class 2: Between Threshold 1 and 2
        mask2 = (sample['reflectance'] >= th1) & (sample['reflectance'] < th2)
        # Class 3: > Threshold 2
        mask3 = sample['reflectance'] >= th2
        
        ax2.scatter(sample.loc[mask1, 'range_3d'], sample.loc[mask1, 'reflectance'], 
                    s=5, alpha=0.5, c='red', label='Class 1 (Low)')
        ax2.scatter(sample.loc[mask2, 'range_3d'], sample.loc[mask2, 'reflectance'], 
                    s=5, alpha=0.5, c='green', label='Class 2 (Mid)')
        ax2.scatter(sample.loc[mask3, 'range_3d'], sample.loc[mask3, 'reflectance'], 
                    s=5, alpha=0.5, c='blue', label='Class 3 (High)')
        
        ax2.axhline(th1, color='black', linestyle='-.')
        ax2.axhline(th2, color='black', linestyle='-.')
        
    ax2.legend()
    plt.tight_layout()
    plt.show()
    
    return splits

# 运行代码
splits = visualize_distributions_trimodal(df)
```

## 2. 高效读取与降采样
使用均匀间隔采样

```python
def task_good_read_and_downsample(file_path, voxel_step=5):
    """
    读取并进行简单的均匀降采样
    :param voxel_step: 每隔多少个点取一个 (降采样率)
    """
    # 1. 读取原始数据
    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    
    # 2. 均匀降采样 (Slicing)
    # [::step] 是 Python 切片的标准用法，效率极高，无需循环
    points_downsampled = points[::voxel_step, :]
    
    return points_downsampled

# --- 测试运行 ---
# 降采样率设为 10 (即保留 10% 的点)
points_small = task_good_read_and_downsample(test_file, voxel_step=10)

print(f"\n【良好】降采样完成 (1/10)。")
print(f"原始点数: {points.shape[0]} -> 降采样后: {points_small.shape[0]}")
```

## 3. 封装接口与去噪
LiDARLoader 可以设置有效范围、降采样率


```python
class LiDARLoader:
    """
    封装的雷达数据加载器，包含距离过滤逻辑
    """
    def __init__(self, base_dir):
        self.velo_dir = os.path.join(base_dir, 'velodyne')
        
    def read_frame(self, frame_id, downsample=1,
                    enable_distance_filter=True,
                    min_dist=2.0, max_dist=120.0):
        """
        读取指定帧的点云，并根据距离进行过滤
        
        :param frame_id: 帧号，如 0, 1, 100
        :param downsample: 降采样倍率
        :param enable_distance_filter: 是否启用距离区间过滤
        :param min_dist: 最小保留距离（米）
        :param max_dist: 最大保留距离（米）
        """
        # 1. 构造文件路径
        file_name = f"{frame_id:06d}.bin"
        file_path = os.path.join(self.velo_dir, file_name)
        
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File not found: {file_path}")
            
        # 2. 读取点云 (N × 4: x, y, z, intensity)
        points = np.fromfile(file_path, dtype=np.float32).reshape((-1, 4))
        
        # 3. 距离过滤（近处 + 远处）
        if enable_distance_filter:
            dist_sq = points[:, 0]**2 + points[:, 1]**2
            mask = (dist_sq > min_dist**2) & (dist_sq < max_dist**2)
            points = points[mask]
            
        # 4. 降采样
        if downsample > 1:
            points = points[::downsample]
        
        return points


# --- 测试运行 ---
loader = LiDARLoader(DATA_PATH)

# 读取第 50 帧，去掉车身周围 2.5 米内的点 (防车身自遮挡)，保留 50% 的点
cleaned_points = loader.read_frame(50, downsample=2, min_dist=2.5)

print(f"\n【优秀】封装读取第 50 帧完成。")
print(f"清洗后数据形状: {cleaned_points.shape}")
print(f"数据列含义: [x (前), y (左), z (上), intensity]")

```
