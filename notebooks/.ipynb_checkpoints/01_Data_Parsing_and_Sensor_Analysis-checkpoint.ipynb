## 1. 读取二进制数据与数据分析

### a. 数据分布

读取、分析与可视化都是00000.bin帧的数据

```python
def task_qualified_read(file_path):
    """
    读取 KITTI .bin 文件，返回原始点云矩阵
    """
    # 1. 使用 numpy 的 fromfile 读取二进制流
    # KITTI 数据全部是 float32 格式 [cite: 59]
    scan = np.fromfile(file_path, dtype=np.float32)
    
    # 2. 重塑矩阵：每4个数据代表一个点 (x, y, z, intensity) [cite: 61]
    # reshape(-1, 4) 会自动计算点的数量 N
    points = scan.reshape((-1, 4))
    
    return points

# --- 测试运行 ---
test_file = os.path.join(VELODYNE_PATH, '000000.bin')
points = task_qualified_read(test_file)

print(f"读取完成。")
print(f"点云形状: {points.shape}") # 应该看到类似 (12xxxx, 4)
print(f"前5个点数据:\n{points[:5, :]}")
```

```python
import pandas as pd
import numpy as np

def analyze_stats(points):
    """
    计算点云的统计特性
    输入: points (N, 4) 矩阵
    """
    # 将数据转换为 DataFrame 方便查看，命名列
    df = pd.DataFrame(points, columns=['x', 'y', 'z', 'reflectance'])
    
    # 计算距离 (Range)，这是雷达很重要的属性
    df['range_2d'] = np.sqrt(df['x']**2 + df['y']**2)
    df['range_3d'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)
    
    print("====== 点云基础统计信息 ======")
    print(df.describe().round(3)) # 保留3位小数
    
    print("\n====== Reflectance 零值分析 ======")
    # 1. 严格等于 0 的点
    exact_zero_count = (df['reflectance'] == 0).sum()
    print(f"Reflectance == 0 的点数: {exact_zero_count}")
    
    # 2. 近似等于 0 的点 (处理浮点数精度，或者非常小的噪声值)
    # 这里的阈值设为 1e-6，你可以根据需要调整，例如之前可视化的 0.001
    epsilon = 1e-6
    approx_zero_count = (df['reflectance'] <= epsilon).sum()
    print(f"Reflectance ≈ 0 (<= {epsilon}) 的点数: {approx_zero_count}")
    
    return df

# --- 使用示例 ---
df = analyze_stats(points)
```

### b. 雷达采样顺序

验证可得，雷达是顺时针旋转，从正前方x轴正方向开始。  

雷达转一圈代表代表一个倾角的雷达扫描出的结果，从高到低顺序有64个不同倾角。

这一帧数据似乎代表雷达转一圈的结果。每个角度转一圈扫描约1916个点。

```python
import numpy as np
import matplotlib.pyplot as plt
import os

def visualize_azimuth_and_z_with_cycles(file_path):
    """
    1. 计算并打印角度周期数。
    2. 在同一张图上对比 Azimuth (角度) 和 Z (高度)。
    """
    # 1. 读取数据
    if not os.path.exists(file_path):
        print(f"Error: File not found at {file_path}")
        return

    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    
    # 2. 提取数据
    x = points[:, 0]
    y = points[:, 1]
    z = points[:, 2]
    # arctan2 范围是 [-pi, pi]
    azimuth = np.arctan2(y, x)
    
    indices = np.arange(len(azimuth))

    # =======================================================
    # 新增功能：计算角度周期 (Cycle Counting)
    # =======================================================
    # 计算相邻点的角度差
    diffs = np.diff(azimuth)
    
    # 检测“跳变”：当角度从 3.14 变成 -3.14 时，差值绝对值约为 6.28
    # 我们设定阈值为 5.0 (略小于 2*pi) 来检测这种 wrap-around
    cycle_jumps = np.sum(np.abs(diffs) > 5.0)
    
    # 使用 unwrap 计算累积的总旋转角度（消除锯齿，变成直线）
    azimuth_unwrapped = np.unwrap(azimuth)
    total_rotation_rad = azimuth_unwrapped[-1] - azimuth_unwrapped[0]
    total_cycles_precise = abs(total_rotation_rad) / (2 * np.pi)

    print("=" * 40)
    print(f"数据点总数 (Total Points): {len(azimuth)}")
    print(f"检测到的周期跳变次数 (Sawtooth Resets): {cycle_jumps}")
    print(f"总累计旋转角度: {total_rotation_rad:.2f} rad")
    print(f"精确估算旋转圈数: {total_cycles_precise:.4f} 圈")
    print("=" * 40)

    # =======================================================
    # 3. 创建画布
    # =======================================================
    fig, (ax_macro, ax_micro) = plt.subplots(1, 2, figsize=(18, 7))
    
    # -------------------------------------------------------
    # 图 1：宏观视角 (Macro View - 全量数据)
    # -------------------------------------------------------
    limit = int(len(azimuth) * 1.0) # 使用全部数据
    
    # --- 左轴：Azimuth (蓝色) ---
    color_azi = 'tab:blue'
    ax_macro.set_xlabel('Point Index (Time Order)')
    ax_macro.set_ylabel('Azimuth (rad)', color=color_azi, fontsize=12)
    ax_macro.plot(indices[:limit], azimuth[:limit], '.', markersize=0.5, color=color_azi, alpha=0.5)
    ax_macro.tick_params(axis='y', labelcolor=color_azi)
    
    # 标题增加周期信息
    ax_macro.set_title(f"Macro View (All {limit} points)\nCycles Detected: {cycle_jumps} | Total Rotation: {total_cycles_precise:.2f} turns")
    ax_macro.grid(True, axis='x') 

    # --- 右轴：Z (红色) ---
    ax_macro_z = ax_macro.twinx()  
    color_z = 'tab:red'
    ax_macro_z.set_ylabel('Z Coordinate (meters)', color=color_z, fontsize=12)
    
    # 保持你要求的样式：大点，高透明度，颜色深
    ax_macro_z.plot(indices[:limit], z[:limit], '.', markersize=1.5, color=color_z, alpha=0.6)
    
    ax_macro_z.tick_params(axis='y', labelcolor=color_z)
    ax_macro_z.set_ylim(-3, 1) 

    # -------------------------------------------------------
    # 图 2：微观视角 (Micro View - 前 200 个点)
    # -------------------------------------------------------
    micro_limit = 200
    
    # --- 左轴：Azimuth (蓝色) ---
    ax_micro.set_xlabel('Point Index (Time Order)')
    ax_micro.set_ylabel('Azimuth (rad)', color=color_azi, fontsize=12)
    ax_micro.plot(indices[:micro_limit], azimuth[:micro_limit], '.-', 
                  linewidth=1.5, markersize=4, color=color_azi, label='Azimuth')
    ax_micro.tick_params(axis='y', labelcolor=color_azi)
    ax_micro.set_title(f"Micro View (First {micro_limit} points)\nStep (Firing Block) vs. Oscillation")
    ax_micro.grid(True)

    # --- 右轴：Z (红色) ---
    ax_micro_z = ax_micro.twinx()
    ax_micro_z.plot(indices[:micro_limit], z[:micro_limit], '-x', 
                    linewidth=0.8, markersize=4, color=color_z, alpha=0.8, label='Z Coord')
    ax_micro_z.tick_params(axis='y', labelcolor=color_z)
    
    plt.tight_layout()
    plt.show()

# --- 运行部分 ---
# 请替换为你的文件路径
# test_file = os.path.join(VELODYNE_PATH, '000000.bin') 
visualize_azimuth_and_z_with_cycles(test_file)
```

计算出角度突变的索引，从而计算每个倾角对应的点的数目。从结果看，它们并不一致。

```python
import numpy as np
import matplotlib.pyplot as plt
import os

def analyze_points_per_ring(file_path):
    """
    1. 计算方位角突变（相位卷绕）的位置索引。
    2. 根据索引分割数据，统计每一圈（Ring）的点数。
    3. 绘制条形统计图。
    """
    if not os.path.exists(file_path):
        print(f"Error: File not found at {file_path}")
        return

    # 1. 读取数据
    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    
    # 2. 计算方位角
    x = points[:, 0]
    y = points[:, 1]
    azimuth = np.arctan2(y, x)
    
    # 3. 寻找突变点 (Wrap-around indices)
    # 计算相邻角度差
    diffs = np.diff(azimuth)
    
    # 阈值设为 5.0 (接近 2*pi)，找到突变位置
    # np.where 返回的是满足条件的下标元组，[0] 取出数组
    # +1 是因为 diff[i] 是 arr[i+1] - arr[i]，突变发生在 i 和 i+1 之间，
    # 所以新的一圈是从 i+1 开始的
    split_indices = np.where(np.abs(diffs) > 5.0)[0] + 1
    
    # 4. 计算每一圈的点数
    ring_counts = []
    
    # 添加起始点 0 和 结束点 len(points) 以便循环计算
    boundaries = [0] + list(split_indices) + [len(points)]
    
    for i in range(len(boundaries) - 1):
        start = boundaries[i]
        end = boundaries[i+1]
        count = end - start
        ring_counts.append(count)
    
    # 5. 打印统计信息
    num_rings = len(ring_counts)
    avg_points = np.mean(ring_counts)
    std_points = np.std(ring_counts)
    
    print("=" * 40)
    print(f"检测到的完整/部分圈数 (Rings): {num_rings}")
    print(f"每圈平均点数: {avg_points:.2f}")
    print(f"每圈点数标准差: {std_points:.2f}")
    print(f"最少点数: {np.min(ring_counts)} (Ring {np.argmin(ring_counts)})")
    print(f"最多点数: {np.max(ring_counts)} (Ring {np.argmax(ring_counts)})")
    print("=" * 40)
    
    # 6. 绘制条形图
    plt.figure(figsize=(16, 6))
    
    bars = plt.bar(range(num_rings), ring_counts, color='skyblue', edgecolor='black', alpha=0.7)
    
    # 绘制平均线
    plt.axhline(y=avg_points, color='r', linestyle='--', linewidth=2, label=f'Average: {int(avg_points)}')
    
    plt.title(f"Number of Points per Ring (Total {num_rings} Rings Detected)")
    plt.xlabel("Ring Index (0 = Top/Bottom most beam)")
    plt.ylabel("Point Count")
    plt.legend()
    plt.grid(True, axis='y', linestyle='--', alpha=0.5)
    
    # 如果圈数不是特别多，可以在柱子上标数字
    if num_rings <= 70:
        for i, bar in enumerate(bars):
            height = bar.get_height()
            # 只标记异常值或者每隔几个标记一下，避免太拥挤
            if abs(height - avg_points) > std_points * 2 or i % 5 == 0:
                plt.text(bar.get_x() + bar.get_width()/2., height,
                         f'{int(height)}',
                         ha='center', va='bottom', fontsize=8, rotation=90)

    plt.tight_layout()
    plt.show()
    
    return ring_counts, split_indices

# --- 运行 ---
# 请替换为你的文件路径
test_file = os.path.join(VELODYNE_PATH, '000000.bin') 
counts, indices = analyze_points_per_ring(test_file)
```

计算不同帧的突变索引是否相同，很遗憾也不相同。

```python
import numpy as np
import matplotlib.pyplot as plt
import os

def get_split_indices(file_path):
    """
    辅助函数：读取单个文件并返回突变索引
    """
    if not os.path.exists(file_path):
        return None
    
    # 读取和计算
    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    x = points[:, 0]
    y = points[:, 1]
    azimuth = np.arctan2(y, x)
    
    # 找突变
    diffs = np.diff(azimuth)
    split_indices = np.where(np.abs(diffs) > 5.0)[0] + 1
    
    return split_indices

def analyze_multi_frame_stability(base_path, start_frame=0, num_frames=5):
    """
    对比多帧数据的 Ring 分割点位置稳定性
    """
    indices_list = []
    valid_frames = []

    print(f"正在分析从 {start_frame} 开始的 {num_frames} 帧数据...")

    # 1. 循环读取 5 帧
    for i in range(start_frame, start_frame + num_frames):
        # 构造文件名 000000.bin, 000001.bin ...
        file_name = f'{i:06d}.bin'
        file_path = os.path.join(base_path, file_name)
        
        idxs = get_split_indices(file_path)
        
        if idxs is not None:
            indices_list.append(idxs)
            valid_frames.append(i)
            print(f"Frame {i}: 找到 {len(idxs)} 个分割点")
        else:
            print(f"Frame {i}: 文件不存在，跳过。")

    if len(indices_list) < 2:
        print("有效帧数不足，无法进行对比。")
        return

    # 2. 可视化对比 (Visual Check)
    plt.figure(figsize=(15, 8))
    
    # 使用散点图：X轴是索引位置，Y轴是帧ID
    # 如果位置相同，点应该会在垂直方向完美对齐，形成竖直线
    colors = plt.cm.viridis(np.linspace(0, 1, len(indices_list)))
    
    for i, (frame_id, idxs) in enumerate(zip(valid_frames, indices_list)):
        plt.scatter(idxs, np.full_like(idxs, frame_id), 
                    s=20, c=[colors[i]], marker='|', label=f'Frame {frame_id}')

    plt.title(f"Split Indices Alignment Across {len(valid_frames)} Frames")
    plt.xlabel("Point Index (Split Position)")
    plt.ylabel("Frame ID")
    plt.yticks(valid_frames)
    plt.grid(True, axis='x', linestyle='--', alpha=0.5)
    
    # 为了看清细节，我们只显示前 5 个分割点的局部放大图会更有意义，但这里先画全图
    # 你可以使用 plt.xlim(0, 5000) 来放大查看前两圈
    plt.tight_layout()
    plt.show()

    # 3. 数值稳定性验证 (Numerical Verification)
    print("\n" + "="*40)
    print("      突变位置稳定性分析报告")
    print("="*40)
    
    base_idxs = indices_list[0]
    base_frame = valid_frames[0]
    
    is_universal = True
    max_drift_total = 0
    
    for i in range(1, len(indices_list)):
        curr_idxs = indices_list[i]
        curr_frame = valid_frames[i]
        
        # 3.1 检查数量是否一致
        if len(base_idxs) != len(curr_idxs):
            print(f"[警告] Frame {curr_frame} 的圈数 ({len(curr_idxs)}) 与 Frame {base_frame} ({len(base_idxs)}) 不一致！")
            is_universal = False
            continue
            
        # 3.2 检查位置偏差
        diff = curr_idxs - base_idxs
        max_drift = np.max(np.abs(diff))
        mean_drift = np.mean(np.abs(diff))
        max_drift_total = max(max_drift_total, max_drift)
        
        print(f"Frame {base_frame} vs {curr_frame}:")
        print(f"  - 最大索引偏差: {max_drift} 个点")
        print(f"  - 平均索引偏差: {mean_drift:.2f} 个点")
        
        if max_drift > 0:
            is_universal = False
            
    print("-" * 40)
    if is_universal:
        print("【结论】 完美！所有帧的突变位置完全一致。")
        print("       -> 可以使用静态索引表 (Static Lookup Table)。")
    elif max_drift_total < 5:
        print(f"【结论】 基本一致。最大偏差仅为 {max_drift_total} 个点。")
        print("       -> 雷达转速非常稳定，误差通常来自数据包边界的微小抖动。")
        print("       -> 依然可以用静态表，但建议加上 +/- 5 的容差范围。")
    else:
        print(f"【结论】 不通用。最大偏差达到 {max_drift_total} 个点。")
        print("       -> 雷达转速有波动，或者每圈点数不固定。")
        print("       -> 必须对每一帧单独动态计算分割索引。")
    print("=" * 40)

# --- 运行 ---
# 请确保 VELODYNE_PATH 变量已定义
analyze_multi_frame_stability(VELODYNE_PATH, start_frame=0, num_frames=5)
```

## 2. 数据分布可视化

### a.地面高度分析

观察z=-1.73(理想地面)为界限的数据分布是否有明显不同。结果证明，没有。

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

# 正常高度的颜色渐进 colormap（你也可以换）
KITTICMAP = LinearSegmentedColormap.from_list(
    "kitti_height", ["#0000ff", "#00ff00", "#ff0000"]
)

def visualize_spatial(df, z_split=-1.73):
    """
    左图（BEV）：
        - 空白背景 = 黑色
        - z < z_split = 浅蓝色
        - z >= z_split = 灰度随 z 增强变亮
    右图：
        - 比较 z < z_split 与 z >= z_split 的 reflectance 直方图分布
    """

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    # ============================================================
    # 1. 左图 BEV
    # ============================================================
    x = df['x'].values
    y = df['y'].values
    z = df['z'].values

    z_min = z.min()
    z_max = z.max()

    colors = np.zeros((len(df), 4))  # RGBA

    # --- ① 阈值以下：浅蓝色 ---
    low_mask = (z < z_split)
    colors[low_mask] = [0.4, 0.6, 1.0, 1.0]

    # --- ② 阈值以上：亮度随高度增强 ---
    high_mask = (z >= z_split)
    z_norm = (z[high_mask] - z_split) / (z_max - z_split)
    z_norm = np.clip(z_norm, 0, 1)

    min_gray = 0.3
    gray_vals = min_gray + z_norm * (1.0 - min_gray)  # 0.3 ~ 1.0 之间

    colors[high_mask] = np.stack([gray_vals, gray_vals, gray_vals, np.ones_like(gray_vals)], axis=1)

    # 设置背景为黑色
    ax1.set_facecolor("black")

    ax1.scatter(x, y, c=colors, s=0.1)
    ax1.axis("equal")
    ax1.set_title("Bird's Eye View (Custom Color Scheme)")
    ax1.set_xlabel("X (Forward)")
    ax1.set_ylabel("Y (Left/Right)")

    # ============================================================
    # 2. 右图 Reflectance 直方图对比
    # ============================================================

    refl_low = df.loc[df['z'] < z_split, 'reflectance']
    refl_high = df.loc[df['z'] >= z_split, 'reflectance']

    bins = 50

    ax2.hist(
        refl_low,
        bins=bins,
        alpha=0.6,
        color='skyblue',
        label=f"z < {z_split:.2f}",
        density=False
    )
    ax2.hist(
        refl_high,
        bins=bins,
        alpha=0.6,
        color='white',
        edgecolor='gray',
        label=f"z ≥ {z_split:.2f}",
        density=False
    )

    ax2.set_title("Reflectance Distribution Comparison")
    ax2.set_xlabel("Reflectance")
    ax2.set_ylabel("Density")
    ax2.legend()

    plt.show()


# --- 运行可视化 ---
visualize_spatial(df, z_split=-1.73)
```

### b. 地面坡度分析

### c. 强度的分布分析

强度分布为双峰分布， reflectance=0 有很多样本

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_distributions_manual(df, vline1_pos=0.25, vline2_pos=0.35, n_bins_main=60, n_bins_zoom=50):
    """
    可视化强度分布直方图。
    图1：整体分布图，并绘制两条可调整位置的竖线。
    图2：聚焦于 0 到 0.03 范围内的分布图。
    
    参数:
    - df: 包含 'reflectance' 列的 DataFrame。
    - vline1_pos: 第一条竖线的位置 (默认 0.25)。
    - vline2_pos: 第二条竖线的位置 (默认 0.35)。
    - n_bins_main: 主图的直方图 bin 数量。
    - n_bins_zoom: 放大图的直方图 bin 数量。
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    refl = df['reflectance'].values
    
    # =========================
    # 1. 主直方图 + 手动竖线
    # =========================
    # 绘制整体分布直方图
    ax1.hist(refl, bins=n_bins_main, log=False, color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_title("Reflectance Distribution (Main View)")
    ax1.set_xlabel("Reflectance")
    ax1.set_ylabel("Count")
    
    # 绘制指定的两条竖线
    # 使用不同的颜色和线型来区分
    ax1.axvline(x=vline1_pos, color='red', linestyle='--', linewidth=2, label=f'Manual Split 1: {vline1_pos}')
    ax1.axvline(x=vline2_pos, color='green', linestyle='-.', linewidth=2, label=f'Manual Split 2: {vline2_pos}')
    
    # 添加图例以显示竖线的值
    ax1.legend()
    
    # =========================
    # 2. 聚焦直方图 (0 - 0.03 范围)
    # =========================
    # 定义聚焦范围
    zoom_min, zoom_max = 0.0, 0.03
    
    # 筛选出该范围内的数据
    refl_zoom = refl[(refl >= zoom_min) & (refl <= zoom_max)]
    
    if len(refl_zoom) > 0:
        # 绘制放大后的直方图
        # 使用较细的 bin 来展示该狭窄范围内的细节
        ax2.hist(refl_zoom, bins=n_bins_zoom, color='orange', edgecolor='black', alpha=0.7)
    else:
        print(f"Warning: No data found in range [{zoom_min}, {zoom_max}]")
        
    ax2.set_title(f"Reflectance Distribution (Zoomed: {zoom_min}-{zoom_max})")
    ax2.set_xlabel("Reflectance")
    ax2.set_ylabel("Count")
    
    # 强制设置X轴范围，确保视图聚焦
    ax2.set_xlim(zoom_min, zoom_max)
    # 添加网格方便查看
    ax2.grid(True, axis='x', linestyle='--', alpha=0.5)
    
    plt.tight_layout()
    plt.show()


# 使用默认参数运行 (竖线在 0.25 和 0.35)
print("Displaying with custom split lines (0.25, 0.35)...")
visualize_distributions_manual(df, vline1_pos=0.25, vline2_pos=0.35)
```

reflectance=0 看起来不应该舍弃，数据并没有明显问题；除了小车内部似乎有些噪声。

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_reflectance_subsets(df, range_min=0.25, range_max=0.35, z_split=-1.73, point_size=0.5):
    """
    可视化两类特定反射率的点云分布（鸟瞰图 BEV）。
    两张图均遵循统一的视觉标准：
        - 背景 = 黑色
        - z < z_split = 浅蓝色 (地面)
        - z >= z_split = 灰度随 z 增强变亮 (障碍物)
    
    参数:
    - df: 包含 'x', 'y', 'z', 'reflectance' 的 DataFrame
    - range_min, range_max: 右图筛选反射率的范围
    - z_split: 地面与障碍物的高度分割阈值 (默认 -1.73)
    - point_size: 散点大小
    """
    
    # 获取全局最大高度，保证两张图的灰度亮度标准一致
    global_z_max = df['z'].max()
    
    # 定义一个内部辅助函数来生成颜色
    def get_colors(z_values):
        colors = np.zeros((len(z_values), 4)) # RGBA
        
        # --- ① 阈值以下：浅蓝色 ---
        low_mask = (z_values < z_split)
        colors[low_mask] = [0.4, 0.6, 1.0, 1.0]
        
        # --- ② 阈值以上：亮度随高度增强 ---
        high_mask = (z_values >= z_split)
        if np.any(high_mask):
            # 使用全局最大高度进行归一化，确保亮度在不同图中具有可比性
            denom = max(0.1, global_z_max - z_split) # 防止除以0
            z_norm = (z_values[high_mask] - z_split) / denom
            z_norm = np.clip(z_norm, 0, 1)
            
            min_gray = 0.3
            gray_vals = min_gray + z_norm * (1.0 - min_gray)
            
            # 赋值 RGBA (灰度 r=g=b)
            colors[high_mask] = np.stack([gray_vals, gray_vals, gray_vals, np.ones_like(gray_vals)], axis=1)
            
        return colors

    # 创建画布
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    # 提取原始数据
    x = df['x'].values
    y = df['y'].values
    z = df['z'].values
    refl = df['reflectance'].values

    # ============================================================
    # 1. 左图：只显示 Reflectance ≈ 0
    # ============================================================
    mask_zero = (refl <= 0.001)
    
    # 设置黑色背景
    ax1.set_facecolor("black")
    
    if np.any(mask_zero):
        x_sub = x[mask_zero]
        y_sub = y[mask_zero]
        z_sub = z[mask_zero]
        
        # 获取符合要求的颜色
        c_sub = get_colors(z_sub)
        
        ax1.scatter(x_sub, y_sub, c=c_sub, s=point_size)
        count_zero = np.sum(mask_zero)
    else:
        count_zero = 0
        
    ax1.axis("equal")
    ax1.set_title(f"Subset 1: Reflectance ≈ 0 (Count: {count_zero})\n(z < {z_split} Blue, z >= {z_split} Gray)")
    ax1.set_xlabel("X")
    ax1.set_ylabel("Y")
    ax1.grid(False)

    # ============================================================
    # 2. 右图：显示 Reflectance 在 [range_min, range_max] 之间
    # ============================================================
    mask_range = (refl >= range_min) & (refl <= range_max)
    
    # 设置黑色背景
    ax2.set_facecolor("black")
    
    if np.any(mask_range):
        x_sub = x[mask_range]
        y_sub = y[mask_range]
        z_sub = z[mask_range]
        
        # 获取符合要求的颜色
        c_sub = get_colors(z_sub)
        
        ax2.scatter(x_sub, y_sub, c=c_sub, s=point_size)
        count_range = np.sum(mask_range)
    else:
        count_range = 0
        
    ax2.axis("equal")
    ax2.set_title(f"Subset 2: Reflectance in [{range_min}, {range_max}] (Count: {count_range})\n(z < {z_split} Blue, z >= {z_split} Gray)")
    ax2.set_xlabel("X")
    ax2.set_ylabel("Y")
    ax2.grid(False)

    plt.tight_layout()
    plt.show()

# ==========================================
# 运行示例
# ==========================================
# 假设 df 已经存在
# z_split 默认为 -1.73，你可以根据需要调整
visualize_reflectance_subsets(df, range_min=0.25, range_max=0.35, z_split=-1.73)
```

reflectance和角度、距离的关系

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_distributions_trimodal(df, n_bins=60, sample_n=10000):
    """
    可视化反射率与：
    1. 方位角(angle)
    2. range_2d
    3. range_3d
    的关系（三张图并排）。

    注意：本版本不再做三峰分析，也不再计算分割阈值。
    """

    # ========= 准备数据 =========
    # 如果还没有 range_2d / range_3d，就自动计算一下，方便独立使用
    if 'range_2d' not in df.columns or 'range_3d' not in df.columns:
        if not {'x', 'y', 'z'}.issubset(df.columns):
            raise ValueError("df 中缺少 x/y/z 列，无法自动计算 range_2d / range_3d")
        df = df.copy()
        df['range_2d'] = np.sqrt(df['x']**2 + df['y']**2)
        df['range_3d'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)

    # 计算方位角（单位：度）
    # atan2(y, x)：原点指向点 (x, y) 的水平夹角
    df = df.copy()
    df['angle_deg'] = np.degrees(np.arctan2(df['y'], df['x']))

    # 抽样，避免点太多画图太慢
    if len(df) > sample_n:
        sample = df.sample(n=sample_n, random_state=42)
    else:
        sample = df

    # ========= 画三张图：并排 =========
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Reflectance vs Angle
    ax1.scatter(sample['angle_deg'], sample['reflectance'], s=3, alpha=0.5)
    ax1.set_title("Reflectance vs Angle")
    ax1.set_xlabel("Angle (deg)")
    ax1.set_ylabel("Reflectance")
    ax1.grid(True, linestyle='--', alpha=0.3)

    # 2. Reflectance vs Range-2D
    ax2.scatter(sample['range_2d'], sample['reflectance'], s=3, alpha=0.5)
    ax2.set_title("Reflectance vs Range-2D")
    ax2.set_xlabel("Range-2D (m)")
    ax2.set_ylabel("Reflectance")
    ax2.grid(True, linestyle='--', alpha=0.3)

    # 3. Reflectance vs Range-3D
    ax3.scatter(sample['range_3d'], sample['reflectance'], s=3, alpha=0.5)
    ax3.set_title("Reflectance vs Range-3D")
    ax3.set_xlabel("Range-3D (m)")
    ax3.set_ylabel("Reflectance")
    ax3.grid(True, linestyle='--', alpha=0.3)

    plt.tight_layout()
    plt.show()

    # 现在不再返回 splits 了，如有需要可以改成 return df 或直接不返回
    # 这里保持兼容性，返回 None
    return None

# 运行代码
splits = visualize_distributions_trimodal(df)
```

### d. 距离分布的分析

从分布图看，小于3m大于50m的基本可以不要。

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_range_distribution(df, bins_main=100, bins_zoom=50):
    """
    可视化 Range-3D 的分布情况（直方图）。
    图1：整体分布 (Full Range)
    图2：近距离细节 (0-5m Zoom)
    """
    
    # ========= 1. 数据准备 =========
    # 确保有 range_3d 列
    if 'range_3d' not in df.columns:
        if {'x', 'y', 'z'}.issubset(df.columns):
            df = df.copy() # 避免修改原始 df
            df['range_3d'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)
        else:
            raise ValueError("df 中缺少 x, y, z 列，无法计算 range_3d")
            
    range_vals = df['range_3d'].values
    
    # ========= 2. 绘图 =========
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # --- 左图：全部范围统计 ---
    # 使用较多的 bins 以便看清整体轮廓
    ax1.hist(range_vals, bins=bins_main, color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_title(f"Range-3D Distribution (Full Range: 0 - {range_vals.max():.1f}m)")
    ax1.set_xlabel("Range-3D (m)")
    ax1.set_ylabel("Count")
    ax1.grid(True, axis='y', linestyle='--', alpha=0.5)
    
    # --- 右图：0-5m 细节统计 ---
    # 筛选数据
    mask_zoom = (range_vals >= 0) & (range_vals <= 5)
    zoom_data = range_vals[mask_zoom]
    
    if len(zoom_data) > 0:
        # 绘制直方图
        ax2.hist(zoom_data, bins=bins_zoom, color='orange', edgecolor='black', alpha=0.7)
        ax2.set_title(f"Range-3D Distribution (Zoom: 0-5m)\nCount: {len(zoom_data)}")
    else:
        ax2.text(0.5, 0.5, "No points in 0-5m range", ha='center', va='center')
        ax2.set_title("Range-3D Distribution (Zoom: 0-5m)")
        
    ax2.set_xlabel("Range-3D (m)")
    ax2.set_ylabel("Count")
    ax2.set_xlim(0, 5) # 强制锁定 X 轴范围
    ax2.grid(True, axis='x', linestyle='--', alpha=0.5)
    ax2.grid(True, axis='y', linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 运行示例
# ==========================================
# 假设 df 已经存在
visualize_range_distribution(df)
```

## 2. 高效读取与降采样
使用均匀间隔采样

```python
def task_good_read_and_downsample(file_path, voxel_step=5):
    """
    读取并进行简单的均匀降采样
    :param voxel_step: 每隔多少个点取一个 (降采样率)
    """
    # 1. 读取原始数据
    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    
    # 2. 均匀降采样 (Slicing)
    # [::step] 是 Python 切片的标准用法，效率极高，无需循环
    points_downsampled = points[::voxel_step, :]
    
    return points_downsampled

# --- 测试运行 ---
# 降采样率设为 10 (即保留 10% 的点)
points_small = task_good_read_and_downsample(test_file, voxel_step=10)

print(f"\n【良好】降采样完成 (1/10)。")
print(f"原始点数: {points.shape[0]} -> 降采样后: {points_small.shape[0]}")
```

```python
import numpy as np
import os

def task_structured_downsample(file_path, vertical_step=2, horizontal_step=5):
    """
    结构化降采样：基于线束(Ring)结构的双重采样。
    
    参数:
    - vertical_step: 垂直采样步长 (每隔多少圈取一圈)。
                     例如 2 表示只取偶数圈 (64线 -> 32线)。
                     1 表示保留所有圈。
    - horizontal_step: 水平采样步长 (每圈内部每隔多少点取一个)。
                       例如 5 表示水平密度变为原来的 1/5。
    """
    if not os.path.exists(file_path):
        print(f"Error: File not found at {file_path}")
        return None

    # 1. 读取原始数据
    scan = np.fromfile(file_path, dtype=np.float32)
    points = scan.reshape((-1, 4))
    
    # 2. 计算分层索引 (Ring Split Indices)
    #    利用前面验证过的逻辑：寻找方位角(Azimuth)的突变点
    x = points[:, 0]
    y = points[:, 1]
    azimuth = np.arctan2(y, x)
    
    diffs = np.diff(azimuth)
    # 找到所有突变点的位置 (每圈的结束点/下一圈的起始点)
    split_indices = np.where(np.abs(diffs) > 5.0)[0] + 1
    
    # 构建完整的边界列表: [0, idx1, idx2, ..., len]
    boundaries = np.concatenate(([0], split_indices, [len(points)]))
    
    # 3. 执行双重降采样
    sampled_chunks = []
    
    total_rings = len(boundaries) - 1
    kept_rings_count = 0
    
    # --- 第一层循环：垂直采样 (针对 Ring) ---
    # range(start, stop, step) 实现了垂直方向的间隔采样
    for i in range(0, total_rings, vertical_step):
        start = boundaries[i]
        end = boundaries[i+1]
        
        # --- 第二层操作：水平采样 (针对 Ring 内的点) ---
        # 使用切片 [start:end:horizontal_step]
        # start:end 选取当前圈的所有点
        # :horizontal_step 在这个范围内每隔 N 个取一个
        chunk = points[start:end:horizontal_step]
        
        sampled_chunks.append(chunk)
        kept_rings_count += 1
        
    # 4. 重新组装数据
    if len(sampled_chunks) > 0:
        points_downsampled = np.concatenate(sampled_chunks, axis=0)
    else:
        points_downsampled = np.zeros((0, 4), dtype=np.float32)

    # 打印统计信息
    print("=" * 40)
    print(f"【结构化降采样报告】")
    print(f"垂直步长 (Vertical Step): {vertical_step} (每 {vertical_step} 圈取 1 圈)")
    print(f"水平步长 (Horiz. Step):   {horizontal_step} (每 {horizontal_step} 点取 1 点)")
    print("-" * 40)
    print(f"原始圈数: {total_rings} -> 保留圈数: {kept_rings_count}")
    print(f"原始点数: {len(points)} -> 降采后点数: {len(points_downsampled)}")
    print(f"压缩比率: {len(points_downsampled) / len(points):.2%}")
    print("=" * 40)
    
    return points_downsampled

# --- 测试运行 ---
# 假设我们要模拟 32线雷达的效果 (vertical_step=2)，且水平分辨率降低一半 (horizontal_step=2)
# test_file = os.path.join(VELODYNE_PATH, '000000.bin') 
points_structured = task_structured_downsample(test_file, vertical_step=2, horizontal_step=5)

# 验证一下形状
print(points_structured.shape)
```

## 3. 封装接口与去噪
LiDARLoader 可以设置有效范围、降采样率


```python
import numpy as np
import os
import matplotlib.pyplot as plt

class LiDARLoader:
    """
    封装的雷达数据加载器，支持多种采样模式
    """
    def __init__(self, base_dir):
        self.velo_dir = os.path.join(base_dir, 'velodyne')
        
    def _detect_ring_boundaries(self, points):
        """
        [私有方法] 基于方位角突变检测线束边界
        """
        x = points[:, 0]
        y = points[:, 1]
        azimuth = np.arctan2(y, x)
        diffs = np.diff(azimuth)
        # 寻找突变点 (Wrap-around indices)
        split_indices = np.where(np.abs(diffs) > 5.0)[0] + 1
        return np.concatenate(([0], split_indices, [len(points)]))

    def read_frame(self, frame_id, 
                   sampling_method='structured', # 'uniform' 或 'structured'
                   uniform_step=1,            # 均匀采样参数
                   struct_vert_step=1,        # 结构化-垂直参数
                   struct_horiz_step=1,       # 结构化-水平参数
                   enable_distance_filter=True,
                   min_dist=3.0, max_dist=50.0):
        """
        读取指定帧的点云，支持灵活的过滤和采样
        """
        # 1. 构造文件路径
        file_name = f"{frame_id:06d}.bin"
        file_path = os.path.join(self.velo_dir, file_name)
        
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File not found: {file_path}")
            
        # 2. 读取点云 (N × 4)
        points = np.fromfile(file_path, dtype=np.float32).reshape((-1, 4))
        
        # 3. 距离过滤 (建议在结构化采样前做，或者采样后做均可，这里选择先过滤减小计算量)
        # 注意：只要不破坏角度的单调性，距离过滤不会影响后续的Ring检测
        if enable_distance_filter:
            dist_sq = points[:, 0]**2 + points[:, 1]**2
            mask = (dist_sq > min_dist**2) & (dist_sq < max_dist**2)
            points = points[mask]
            
        # 4. 执行降采样
        if sampling_method == 'uniform':
            if uniform_step > 1:
                points = points[::uniform_step]
                
        elif sampling_method == 'structured':
            # 仅当需要降采样时才计算
            if struct_vert_step > 1 or struct_horiz_step > 1:
                boundaries = self._detect_ring_boundaries(points)
                sampled_chunks = []
                
                # 遍历每一圈 (Ring)
                # vertical_step 控制跳过多少圈
                for i in range(0, len(boundaries) - 1, struct_vert_step):
                    start = boundaries[i]
                    end = boundaries[i+1]
                    # horizontal_step 控制圈内稀疏度
                    chunk = points[start:end:struct_horiz_step]
                    sampled_chunks.append(chunk)
                
                if len(sampled_chunks) > 0:
                    points = np.concatenate(sampled_chunks, axis=0)
                else:
                    points = np.zeros((0, 4), dtype=np.float32)
        else:
            raise ValueError(f"Unknown sampling method: {sampling_method}")
            
        return points


# 假设 DATA_PATH 已定义
# frame_id = 50 

loader = LiDARLoader(DATA_PATH)

# --- 方案 A: 简单均匀采样 (1/10) ---
# 所有的点混在一起每10个取1个
points_uniform = loader.read_frame(
    50, 
    sampling_method='uniform', 
    uniform_step=10, 
    min_dist=2.5
)

# --- 方案 B: 结构化采样 (垂直1/2 * 水平1/5 = 1/10) ---
# 只取偶数线束，且每条线束变稀疏5倍
points_structured = loader.read_frame(
    50, 
    sampling_method='structured',
    struct_vert_step=2,
    struct_horiz_step=5,
    min_dist=2.5
)

print(f"Uniform 采样后点数: {points_uniform.shape[0]}")
print(f"Structured 采样后点数: {points_structured.shape[0]}")
```
